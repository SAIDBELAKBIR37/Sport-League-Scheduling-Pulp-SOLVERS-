```python
prob = LpProblem("SportHachimi", LpMinimize)
```

```python
from pulp import *
n= 4 
teams=range(1,n+1)

week=range(1,n)
period=range(1, n//2 + 1 )
```

```python

indices = [str(i)+str(j)+str(k)+str(p) for i in range(n) for j in range(n) for k in range(n-1) for p in range(n//2)]

#print(indices)

Xijkp = LpVariable.dicts('x', (teams,teams,week,period) , lowBound=0 , upBound=1, cat='bianry')


print(Xijkp[1][1][1][1])
```

Fixation des contraintes 

```python
#un match par semaine

for k in week:
    for j in teams :
        prob += (lpSum([Xijkp[i][j][k][p] + Xijkp[j][i][k][p] for p in period for i in teams if i!= j ] ) == 1 )
      
```

```python
# un seul match entre i et j 
for i in teams:
    for i in teams : 
      prob += (lpSum([Xijkp[i][j][k][p] + Xijkp[j][i][k][p] for p in period for k in week if i!= j ] ) == 1 )
      
```

```python
for p in period:
    for j in teams : 
      prob += (lpSum([Xijkp[i][j][k][p] + Xijkp[j][i][k][p] for i in teams for k in week ] ) <= 2 )
```

```python
prob.solve()
#etape de résolution 
```
